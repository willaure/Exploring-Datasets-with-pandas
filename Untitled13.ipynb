```python
# Fetch the data where AreaName is 'Africa' and RegName is 'Southern Africa'
southern_africa_data = df_can[(df_can['AreaName'] == 'Africa') & (df_can['RegName'] == 'Southern Africa')]

# Display the dataframe
display(southern_africa_data)

# Count the number of instances
num_instances = len(southern_africa_data)
print(f"\nNumber of instances: {num_instances}")
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
top_5 = df_can.head(5)
top_5
```

```python
# Get the top 3 countries for 2010 immigration
top_2010 = df_can[['DevName', '2010']].sort_values('2010', ascending=False).head(3)

# Display the results
print("Top 3 Countries for Immigration to Canada in 2010:")
display(top_2010)

# Alternative prettier display
print("\nFormatted Results:")
for index, row in top_2010.iterrows():
    print(f"{row['DevName']}: {row['2010']:,} immigrants")
```
